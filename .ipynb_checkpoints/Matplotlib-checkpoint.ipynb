## matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) #在(0, 10)之间切分100段
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y) #绘制折线图
plt.show()
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny) #在一个图形中绘制2条折线
plt.plot(x, cosy, color='red')
plt.show()
```

```python
plt.plot(x, siny) #在一个图形中绘制2条折线
plt.plot(x, cosy, color='red', linestyle='--')
plt.show()
```

```python
plt.plot(x, siny) #在一个图形中绘制2条折线
plt.plot(x, cosy, color='red', linestyle='--')
plt.xlim(-5, 15) #指定x的范围
plt.ylim(0, 1.5) #指定y的范围
plt.show()
```

```python
plt.plot(x, siny) #在一个图形中绘制2条折线
plt.plot(x, cosy, color='red', linestyle='--')
plt.axis([-1, 11, -2, 2])#同时指定x和y的范围
plt.show()
```

```python
plt.plot(x, siny) #在一个图形中绘制2条折线
plt.plot(x, cosy, color='red', linestyle='--')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.show()
```

```python
plt.plot(x, siny, label='sin(x)') 
plt.plot(x, cosy, color='red', linestyle='--', label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend() #将图示，即将label='sin(x)'和label='cos(x)'显示出来
plt.show()
```

```python
plt.plot(x, siny, label='sin(x)') 
plt.plot(x, cosy, color='red', linestyle='--', label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend() #将图示，即将label='sin(x)'和label='cos(x)'显示出来
plt.title('Welcome to the ML world!')
plt.show()
```

## Scatter plot

```python
plt.scatter(x, siny) #绘制散点图
plt.show()
```

```python
plt.scatter(x, siny) #绘制散点图
plt.scatter(x, cosy, color='red')
plt.show()
```
