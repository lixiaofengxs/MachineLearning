```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## 折线图

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle='--')
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle='--')
plt.xlim(-5, 15)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle='--')
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle='--')
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle='--')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x, siny, label="sinx")
plt.plot(x, cosy, color="red", linestyle='--', label="cosx")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend() #显示label="sinx", label="cosx"
plt.show()
```

```python
plt.plot(x, siny, label="sinx")
plt.plot(x, cosy, color="red", linestyle='--', label="cosx")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend() #显示label="sinx", label="cosx"
plt.title("test matplotlib")
plt.show()
```

## 散点图

绘制二维特征时，横纵坐标都是特征。

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color='red')
plt.show()
```
